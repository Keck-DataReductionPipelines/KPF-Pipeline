## Rebuilding the TSDB
This page contains instructions for rebuilding the Time Series Database (TSDB).  Note that all data in the database will be lost (and reingested)!
In Postgres, it can only one using an administrator role (`timeseriesdba`).  Add the password in the code below.

```python
from database.modules.utils.tsdb import TSDB
```

```python
credentials = {"TSDBUSER": 'timeseriesdba', "TSDBPASS": 'ADD_PASSWORD_HERE'}
myDB = TSDB(backend='psql', credentials=credentials)
```

Drop the tables.

```python
myDB.drop_tables()
```

Create new tables.

```python
myDB = TSDB(backend='psql', credentials=credentials)
```

And ingest data (change dates appropriately).

```python
start_date = '20241001'
end_date   = '20251003'
myDB.ingest_dates_to_db(start_date, end_date)
```

Show database status.

```python
myDB.print_db_status()
```
